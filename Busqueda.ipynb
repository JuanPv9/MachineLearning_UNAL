### **Busqueda automática de datos para extracción de coberturas glaciares y encontrar las series de tiempo**

```python
import cubo
import xarray as xr
```

2. Generación de los puntos de coordenadas para la creación de los cubos

```python
# Sierra Nevada de Santa Marta
latSNSM = 10.8250
longSNSM = 73.6722
```

```python
da = cubo.create(
    lat = latSNSM, #latitud del centro
    lon = longSNSM, #longitud del centro
    collection = "sentinel-2-l2a", #selección del STAC (SpacioTemporal Asset Catalogs)
    start_date = "2000-01-01", #fecha de inicio
    end_date = "2023-03-20", #fecha final
    bands = ["B02", "B03", "B04"], #selección de bandas o None
    edge_size = 64, #tamaño de la imagen en píxeles
    resolution = 10 #resolución en píxeles
    )
```

```python
# Mostrar el cubo generado
da
```

3. Generación de una imagen como una serie de tiempo

```python
(da.sel(band=["B04","B03","B02"])/2000).clip(0,1).plot.imshow(col="time",col_wrap = 5)
```
